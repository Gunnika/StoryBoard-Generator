```python
import os
import requests
from bs4 import BeautifulSoup
```

```python
def get_story_links(main_url):
    response = requests.get(main_url)
    soup = BeautifulSoup(response.content, 'html.parser')
    links = soup.find_all('a')
    story_links = [link.get('href') for link in links if 'author' in link.get('href')]
    return story_links
```

```python
def get_story_links_sub(main_url, name):
    response = requests.get(main_url)
    soup = BeautifulSoup(response.content, 'html.parser')
    links = soup.find_all('a')
    story_links = [link.get('href') for link in links if name in link.get('href')]
    return story_links
```

```python
def scrape_story(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    paragraphs = soup.find_all('p')  # Assuming the story text is in <p> tags
    story_text = ' '.join(p.get_text().replace('\n', '') for p in paragraphs)
    return story_text
```

```python
def scrape_story2(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    paragraphs = soup.find_all('p')  # Assuming the story text is in <p> tags
    story_text = '\n\n\t'.join(p.get_text().strip() for p in paragraphs)
    return story_text
```

```python
def scrape_story3(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    paragraphs = soup.find_all('p')  # Assuming the story text is in <p> tags
    story_text = '\\n\\n\\t'.join(p.get_text().strip() for p in paragraphs)
    return story_text
```

```python
def save_story_as_text_file(story_name, story_text, file_directory):

    directory_path = f"{file_directory}/{story_name}"
    if not os.path.exists(directory_path):
        os.makedirs(directory_path)

    file_path = f"{directory_path}/story.txt"
    with open(file_path, "w") as file:
        file.write(story_text)

    return file_path
```

```python
import os
import re

# Regular expressions compiled once for efficiency
regex_pattern_add = re.compile(r'Add .*? to your library')
regex_pattern_featured = re.compile(r'.*? was featured as .*')
# regex_remove_quotes = re.compile(r'\".*?\"')
# Use DOTALL flag to match across multiple lines
regex_remove_quotes = re.compile(r'\".*?\"', re.DOTALL)

# Clean fairy tales dataset
base_path = "./dataset/fairy tale"
output_base_path = "./clean_dataset2/fairy_tale"
os.makedirs(output_base_path, exist_ok=True)

print("Total fairytales: ", len(os.listdir(base_path)))
for i, folder in enumerate(os.listdir(base_path)):
    if not folder.startswith('.DS_Store'):
        print(i, folder)
        if i==52 or i==186 or i==216:
            continue
        folder_path = os.path.join(base_path, folder)
        story_path = os.path.join(folder_path, 'story.txt')
        new_folder = os.path.join(output_base_path, folder)
        os.makedirs(new_folder, exist_ok=True)
        new_file_path = os.path.join(new_folder, 'story.txt')

        with open(story_path, 'r') as file:
            file_contents = file.read()

        # Remove all text within double quotes and replace line breaks with spaces
#         file_contents = regex_remove_quotes.sub('', file_contents).replace('\n', ' ')
        file_contents = regex_remove_quotes.sub('', file_contents)
        file_contents = file_contents.replace('\n', ' ')

        # Replace " with '
        file_contents = file_contents.replace('"', "'")

        # Find positions of various markers and remove the trailing numbers
        marker_positions = [
            m.start() for m in [regex_pattern_add.search(file_contents), 
                                regex_pattern_featured.search(file_contents)] if m
        ] + [file_contents.find("Return to the The"), 
             file_contents.rfind("\n\n\t\n\n\t\n\n\t\n\n\t"), 
             file_contents.rfind("\n\n\t")]
        marker_positions = [pos for pos in marker_positions if pos != -1]
        file_contents = re.sub(r'\s*\d+\.?\d*\s*$', '', file_contents)

        # Trim the content from the earliest marker if any valid position is found
        if marker_positions:
            file_contents = file_contents[:min(marker_positions)]

        # Write the processed content to a new file
        with open(new_file_path, 'w') as new_file:
            new_file.write(file_contents)
```

```python
import os
import re

# Compile the regular expression with DOTALL flag for multiline matching
remove_quotes_pattern = re.compile(r'\".*?\"', re.DOTALL)
regex_pattern = re.compile(r'.*? was featured as .*')

# Clean short story dataset
base_path = "./dataset/short story"
output_base_path = "./clean_dataset2/short story"
os.makedirs(output_base_path, exist_ok=True)

print("Total short stories in dataset:", len(os.listdir(base_path)))

saved_stories_count = 0  # Counter for saved stories

for i, folder in enumerate(os.listdir(base_path)):
    folder_path = os.path.join(base_path, folder)
    if os.path.isdir(folder_path):
        print(f"Processing {i}: {folder}")
        story_path = os.path.join(folder_path, 'story.txt')

        processed_story = ""
        with open(story_path, 'r') as file:
            for line in file:
                # Remove all text within double quotes (including across multiple lines)
                line = remove_quotes_pattern.sub('', line)

                # Perform string replacements
                line = line.replace('\n', ' ').replace('"', "'")

                # Check for markers and perform regex search
                if "Add  A" in line or "Read the next short story" in line or \
                   "Return to the" in line or "\n\n\t\n\n\t\n\n\t\n\n\t" in line or \
                   regex_pattern.search(line):
                    break

                # Remove trailing numbers in each line
                line = re.sub(r'\s*\d+\.?\d*\s*$', '', line)

                processed_story += line

        # Check if the processed story is less than 20 words
        if len(processed_story.split()) >= 20:
            new_folder = os.path.join(output_base_path, folder)
            os.makedirs(new_folder, exist_ok=True)
            new_story_path = os.path.join(new_folder, 'story.txt')
            with open(new_story_path, 'w') as new_file:
                new_file.write(processed_story)
            saved_stories_count += 1  # Increment the counter
        else:
            print(f"Skipped {folder} as the story is less than 20 words.")

print(f"Total stories saved: {saved_stories_count}")
```
